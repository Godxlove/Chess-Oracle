```python


import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = pd.read_csv('chess_dataset.csv')
dataset.drop("increment_code",inplace=True,axis=1)
```

```python
datasetNew=dataset[['victory_status','moves']]

datasetNew.head()
```

```python
datasetNew=datasetNew[datasetNew['victory_status']=='mate']
datasetNew=datasetNew.drop(['victory_status'],axis=1)
datasetNew.head()
                      
```

```python
import numpy as np
moves_df= datasetNew["moves"].str.split(" ",n=222, expand = True)
col_names = ['move{}'.format(i) for i in range(1, 223)]
moves_df.columns=col_names
df = pd.concat([datasetNew, moves_df], axis=1)
df.drop('moves', axis=1, inplace=True)
# df = df.fillna(value=np.nan)
# df=df.fillna(0.0)
df.head(10)
```

```python
df.dtypes
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
df.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=[X_train.shape[1]]))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
for col in df.columns:
    # Check if the column is a string column
    if df[col].dtype == 'object':
        # Create a label encoder for the column
        le = LabelEncoder()
        # Fit the encoder to the column
        df[col] = le.fit_transform(df[col])
        # Convert the column to float32
        df[col] = df[col].astype('float32')
```

```python

import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '1'

# Create some regression data
X_regression = np.expand_dims(np.arange(0, 1000, 5), axis=1)
y_regression = np.expand_dims(np.arange(100, 1100, 5), axis=1)

# Split it into training and test sets
X_reg_train = X_regression[:150]
X_reg_test = X_regression[150:]
y_reg_train = y_regression[:150]
y_reg_test = y_regression[150:]

tf.random.set_seed(42)

# Recreate the model
model_3 = tf.keras.Sequential([
  tf.keras.layers.Dense(100),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1)
])

# Change the loss and metrics of our compiled model
model_3.compile(loss=tf.keras.losses.mae, # change the loss function to be regression-specific
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['mae']) # change the metric to be regression-specific

# Fit the recompiled model
model_3.fit(X_reg_train, y_reg_train, epochs=100)
```

```python
# Evaluate the model using evaluate() and the testing data
loss, accuracy = model_3.evaluate(X_reg_train,y_reg_train)
print('Test accuracy:', accuracy)
```

```python
# Make predictions using predict() and new data
predictions = model_3.predict(X_reg_test)
```
